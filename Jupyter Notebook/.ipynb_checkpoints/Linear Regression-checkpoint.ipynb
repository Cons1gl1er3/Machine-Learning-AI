```python
# Import required library
import numpy as np
import matplotlib.pyplot as plt
```

After import all the needed library, it's time to plug in the training and testing data.

In this notebook, we assume that we can determine the weight of a person base on his/her height.

```python
# Import the input
height = [147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]
weight = [ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]
```

```python
# Modify the input into usable numpy array
X = np.array([height]).T
y = np.array([weight]).T
```

```python
# Plot the data points for visualization
plt.plot(X, y, 'bo')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

As we can see from above, the data points nearly lie on a line, so linear model might perform well.

The expected output is a vector w represent the coeeficient of the line. In order to use the final formula, we need to modify X to X_bar.

```python
X_bar = np.concatenate((X, np.ones((X.shape[0], 1), dtype = np.int8)), axis = 1)
```

The final formular is $(X^T \cdot X)^{-1} \cdot X^T \cdot y$

```python
w = np.linalg.pinv(X_bar.T.dot(X_bar)).dot(X_bar.T).dot(y)
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(145, 185, 2)
y0 = w_0*x0 + w_1
```

```python
# Drawing the fitting line 
plt.plot(x0, y0, 'r')    
plt.plot(X, y, 'bo')
plt.show()
```

```python
x_predict = 165
print('Predict weight for a person with the height of ' + str(x_predict) + ' is: ' + str(int(w_0*x_predict + w_1)))
```
